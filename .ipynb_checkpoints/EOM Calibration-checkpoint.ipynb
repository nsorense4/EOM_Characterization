```python
from scipy.special import jv
```

```python
##Measured Calibration Data
Vin = array([0,100,200,300,400,500,600,700,800, 900,1000,1100,1150,1180]) #Voltage input to EOM
V0 =  array([230,230,230,225,218,210,208,204,210,183,169,158,151,156]) #Amplitude of carrier
V1 = array([0,0,0,4.2,6.9,9.7,13.2,17.5,23.1,25.8,29.1,34.4,35.7,39.6]) #Amplitude of first sideband
```

```python
#set global plotting parameters
linew = 3.0 #line width
#msize = 14 #marker size
axew = 3.0 #axes width

#set the distance to offset the numbers from the ticks
numpad = 15

#set global tick parameters
majw = axew #major tick width
majl = 10.0 #major tick length
minw = axew #minor tick width
minl = 5.0 #minor tick length

#set global font sizes
axefont = 24 #axis label font size
numsize = 28 #axis number font size
legfont = 23 #legend font size
labelfont=12

f, ax1 = plt.subplots(1, 1, figsize=(8,5))
for axis in ['top','bottom','left','right']:
    ax1.spines[axis].set_linewidth(5)
    
ax1.tick_params(axis='x',which='major',width=majw,length=majl,labelsize=numsize,zorder=1,direction='in',pad=numpad,top='off')
ax1.tick_params(axis='x',which='minor',width=minw,length=minl,labelsize=numsize,zorder=1,direction='in',pad=numpad,top='off')
ax1.tick_params(axis='y',which='major',width=majw,length=majl,labelsize=numsize,zorder=1,direction='in',pad=numpad,right='off')
ax1.tick_params(axis='y',which='minor',width=minw,length=minl,labelsize=numsize,zorder=1,direction='in',pad=numpad,right='off')


#Define frequency space
Vpi = 4084
V = linspace(0,Vpi,1000)


plot(V/Vpi,jv(2,V*pi/Vpi)**2,lw=5)
#plot(V/Vpi,jv(0,V*pi/Vpi)**2,c='darkorange',lw=5)
plot(V/Vpi,jv(1,V*pi/Vpi)**2,c='darkorange',lw=5)
plot(V/Vpi,(V*pi/Vpi/2)**2,'k--',lw=5)

plot(Vin/Vpi,(V1/V0[0]),'.',color='darkorange',markeredgewidth=1.5, markeredgecolor='k',linewidth=1,markersize=22,zorder=3)
xlim(-.003,.3)
ylim(-.003,.2)
#savefig("./EOM_calibrationinset.pdf")
#print (V[argmax(jv(1,V*pi/Vpi)**2)])
```
